# STEP 2 — FEATURE ENGINEERING


```python

def parse_bp(bp_str):
    """Extract systolic and diastolic from a string like '100/60'"""
    try:
        parts = str(bp_str).strip().split('/')
        return float(parts[0]), float(parts[1])
    except:
        return np.nan, np.nan

df_book2[['Systolic_BP', 'Diastolic_BP']] = df_book2['Blood_Pressure'].apply(
    lambda x: pd.Series(parse_bp(x))
)
```

```python


gravida_map = {'1st': 1, '2nd': 2, '3rd': 3, '4th': 4, '5th': 5}
df_book2['Gravida_num'] = df_book2['Gravida'].map(gravida_map).fillna(1)
```

```python
# Clean ALL column names at once — removes spaces, fixes encoding issues
df_book2.columns = (
    df_book2.columns
    .str.strip()
    .str.replace(' ', '_')
    .str.replace('\xa0', '_')
)

print(df_book2.columns.tolist())
```

# STEP 5 — PREPARE TRAIN/TEST SPLITS WITH SMOTE


```python

X_mat = df_maternal.drop(columns=['Risk_Label'])
y_mat = df_maternal['Risk_Label']


sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_mat, y_mat)
print("\nAfter SMOTE:", pd.Series(y_res).value_counts())


X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)
print("Train size:", X_train.shape, " | Test size:", X_test.shape)

# Scale features (important for Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

# STEP 6 — TRAIN 3 MODELS


```python
# Model 1: Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42, multi_class='auto')
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_test_scaled)
print("\n===== Logistic Regression =====")
print(classification_report(y_test, y_pred_lr))
```

```python
#  Model 2: Random Forest
rf = RandomForestClassifier(n_estimators=200, max_depth=None,
                             random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("\n===== Random Forest =====")
print(classification_report(y_test, y_pred_rf))
```

```python
# Model 3: XGBoost
xgb_model = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42
)
xgb_model.fit(X_train, y_train,
              eval_set=[(X_test, y_test)],
              verbose=False)
y_pred_xgb = xgb_model.predict(X_test)
print("\n===== XGBoost =====")
print(classification_report(y_test, y_pred_xgb))
```

# STEP 5 — FINAL CLEAN REBUILD + SMOTE + TRAIN

```python
# ─────────────────────────────────────────────
# STEP 5 — FINAL CLEAN REBUILD + SMOTE + TRAIN
# ─────────────────────────────────────────────
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import xgboost as xgb
import joblib
import warnings
warnings.filterwarnings('ignore')

# ── Step 1: Rebuild UCI features with CORRECT 3-class mapping
df_uci = pd.read_csv('matenal health risk.zip')
df_uci = df_uci.drop_duplicates().reset_index(drop=True)

# Normalize UCI labels
df_uci['RiskLevel_clean'] = df_uci['RiskLevel'].str.strip().str.lower().map({
    'high risk': 0,
    'low risk':  1,
    'mid risk':  2,
})
df_uci = df_uci.dropna(subset=['RiskLevel_clean'])
df_uci['Risk_Label'] = df_uci['RiskLevel_clean'].astype(int)

uci_features = df_uci[['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'HeartRate', 'Risk_Label']].copy()
uci_features.columns = ['Age', 'Systolic_BP', 'Diastolic_BP', 'BS', 'Heart_Rate', 'Risk_Label']
# Fix outlier heart rate
median_hr = uci_features['Heart_Rate'].median()
uci_features['Heart_Rate'] = uci_features['Heart_Rate'].replace(7, median_hr)
print("UCI shape:", uci_features.shape)
print(uci_features['Risk_Label'].value_counts())

# ── Step 2: Rebuild Mendeley features
df_mendeley = pd.read_csv('maternal health risk assesment datset.csv', encoding='latin1')
df_mendeley = df_mendeley.dropna(subset=['Risk Level']).reset_index(drop=True)

# Normalize Mendeley labels
df_mendeley['Risk Level'] = df_mendeley['Risk Level'].str.strip().str.lower()
mendeley_risk_map = {
    'high': 0,
    'low':  1,
    'mid':  2,
    'high risk': 0,
    'low risk':  1,
    'mid risk':  2,
}
df_mendeley['Risk_Label'] = df_mendeley['Risk Level'].map(mendeley_risk_map)
df_mendeley = df_mendeley.dropna(subset=['Risk_Label'])
df_mendeley['Risk_Label'] = df_mendeley['Risk_Label'].astype(int)

# Fill missing numeric columns
for col in ['Systolic BP', 'Diastolic', 'BS', 'Heart Rate']:
    if col in df_mendeley.columns:
        df_mendeley[col] = df_mendeley[col].fillna(df_mendeley[col].median())

df_mendeley = df_mendeley[df_mendeley['Age'] <= 70]
df_mendeley['Heart Rate'] = df_mendeley['Heart Rate'].replace(7, df_mendeley['Heart Rate'].median())

mend_features = df_mendeley[['Age', 'Systolic BP', 'Diastolic', 'BS', 'Heart Rate', 'Risk_Label']].copy()
mend_features.columns = ['Age', 'Systolic_BP', 'Diastolic_BP', 'BS', 'Heart_Rate', 'Risk_Label']
print("\nMendeley shape:", mend_features.shape)
print(mend_features['Risk_Label'].value_counts())

# ── Step 3: Combine + clean
df_maternal = pd.concat([uci_features, mend_features], ignore_index=True)
df_maternal = df_maternal[
    (df_maternal['Age'] >= 10) &
    (df_maternal['Age'] <= 60)
].dropna().reset_index(drop=True)

print("\nCombined shape:", df_maternal.shape)
print("Label distribution:")
print(df_maternal['Risk_Label'].value_counts())
print("\nLabel meaning: 0=High Risk, 1=Low Risk, 2=Mid Risk")
```

```python
# ── Step 4: SMOTE + Train/Test Split
X = df_maternal.drop(columns=['Risk_Label'])
y = df_maternal['Risk_Label']

from imblearn.combine import SMOTETomek

sm = SMOTETomek(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
print("After SMOTETomek:", pd.Series(y_res).value_counts())

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)
print("Train:", X_train.shape, " | Test:", X_test.shape)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

#  Train XGBoost

```python
# ── Step 5: Train XGBoost
import xgboost as xgb
xgb_model = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42
)
xgb_model.fit(X_train, y_train,
              eval_set=[(X_test, y_test)],
              verbose=False)

y_pred = xgb_model.predict(X_test)
print("\n===== XGBoost =====")
print(classification_report(y_test, y_pred,
      target_names=['High Risk', 'Low Risk', 'Mid Risk']))
```
